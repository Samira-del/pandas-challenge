```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
Unique_df = purchase_data.drop_duplicates('SN')
Total_Unique_Player = len(Unique_df)
pd.DataFrame({"Total Players": [Total_Unique_Player]})
```

```python
df_Puchasing_data = purchase_data.drop_duplicates('Item ID')
Unique_Item = len(df_Puchasing_data['Item ID'])
Price = df_Puchasing_data['Price']

Purchasing_Analysis = pd.DataFrame({
            "Number of Unique Items": [Unique_Item],
            "Average Price": [Price.mean()],
            "Number of Purchases": [len(purchase_data['Purchase ID'])],
            "Total Revenue": [purchase_data['Price'].sum()]
            })

Purchasing_Analysis
```

```python

Total_gender = Unique_df['Gender'].count()
Female_Player = Unique_df['Gender'].value_counts()['Female']
Male_Player = Unique_df['Gender'].value_counts()['Male']
non_gender_specific_total = Total_gender - Male_Player - Female_Player

Male_percentage =  Male_Player / Total_gender * 100
Female_percentage = Female_Player /float(Total_gender) * 100
Non_gender_Percentage = non_gender_specific_total / float(Total_gender) * 100

Gender_Dic = [
    {"Gender": "Male", "Total Count": Male_Player, "Percentage of Players": Male_percentage},
    {"Gender": "Female", "Total Count": Female_Player, "Percentage of Players": Female_percentage},
    {"Gender": "Other / Non-Disclosed", "Total Count": non_gender_specific_total, "Percentage of Players": Non_gender_Percentage}
]

Gender_Demographics = pd.DataFrame(Gender_Dic)
Gender_Demographics.style.format({"Percentage of Players": "${:.2f}%"})
Gender_Demographics.set_index('Gender', inplace=True)
Gender_Demographics
```

```python
Total_Gender_Purchase_Account = purchase_data['Gender'].count()

Female_Purchase_Account = purchase_data.loc[purchase_data['Gender'] == 'Female']
Male_Purchase_Account = purchase_data.loc[purchase_data['Gender'] == 'Male']

Num_Female = Female_Purchase_Account['Gender'].count()
Num_Male = Male_Purchase_Account['Gender'].count()
Num_non_gender = Total_Gender_Purchase_Account - Num_Male - Num_Female

Total_Ave_Purchase = purchase_data['Price'].mean()
Ave_Purchase_Female = Female_Purchase_Account['Price'].mean()
Ave_Purchase_Male = Male_Purchase_Account['Price'].mean()
Ave_Purchase_Non_Gender = Total_Purchase_Non_Gender / Num_non_gender 

Total_Purchase = purchase_data['Price'].sum()
Total_Purchase_Female = Female_Purchase_Account['Price'].sum()
Total_Purchase_Male = Male_Purchase_Account['Price'].sum()
Total_Purchase_Non_Gender = Total_Purchase - Total_Purchase_Female - Total_Purchase_Male

Female_Avg_Total_Person = Total_Purchase_Female / Female_Player
Male_Avg_Total_Person = Total_Purchase_Male / Male_Player
Non_gender_Avg_Total_Person = Total_Purchase_Non_Gender / non_gender_specific_total


Gender_Purchasing_Analysis_Dic= [
    {"Gender": "Female", "Purchase Count": Num_Female , "Average Purchase Price": Ave_Purchase_Female, "Total Purchase Value": Total_Purchase_Female, "Avg Total Purchase per Person": Female_Avg_Total_Person},
    {"Gender": "Male", "Purchase Count": Num_Male, "Average Purchase Price": Ave_Purchase_Male, "Total Purchase Value": Total_Purchase_Male,  "Avg Total Purchase per Person":Male_Avg_Total_Person},
    {"Gender": "Other / Non-Disclosed", "Purchase Count": Num_non_gender, "Average Purchase Price": Ave_Purchase_Non_Gender, "Total Purchase Value": Total_Purchase_Non_Gender,  "Avg Total Purchase per Person":Non_gender_Avg_Total_Person}
]

Gender_Purchasing_Analysis = pd.DataFrame(Gender_Purchasing_Analysis_Dic)
Gender_Purchasing_Analysis.style.format({"Average Purchase Price": "${:.2f}", "Avg Total Purchase per Person": "${:.2f}"})
Gender_Purchasing_Analysis.set_index('Gender', inplace=True)
Gender_Purchasing_Analysis
```

```python
Group_Item = purchase_data[['Item ID' ,'Item Name','Price']]
Item_Price = Group_Item.groupby(['Item ID' ,'Item Name'])['Price'].mean()
Purchase_count = Group_Item.groupby(['Item ID'])['Item Name'].value_counts()
Total_Purchace_Value = Item_Price * Purchase_count

df_Popular=Purchase_count.to_frame('Purchase_count')
df_Popular['Item Price']=Item_Price.map("${:,.2f}".format)
df_Popular['Total Purchase Value']=Total_Purchace_Value

Sort_Purchase = df_Popular.sort_values('Purchase_count',ascending=False)
Sort_Purchase.head()
```
